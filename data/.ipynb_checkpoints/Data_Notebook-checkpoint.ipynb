```python
import requests
phone_num = []
charlotte_df['Last Name'] = charlotte_df['Last Name'].str.upper()
charlotte_df.rename(columns={'First Name':'First_Name', 'Last Name':'Last_Name'}, inplace=True)
charlotte_df['Address'] = charlotte_df['Address'].str.title()
charlotte_df['Zip'] = charlotte_df['Zip'][:5]
charlotte_df['Zip'] = charlotte_df['Zip'].astype('str')
charlotte_df['Address'] = charlotte_df['Address'].str.replace(r'[^\w\s]+', '')
charlotte_df["Address"]= charlotte_df['Address'].str.replace(' STREET', ' ST')
charlotte_df["Address"]= charlotte_df['Address'].str.replace(' BOULEVARD', ' BLVD')
charlotte_df["Address"]= charlotte_df['Address'].str.replace(' COURT', " CT")
charlotte_df["Address"]= charlotte_df['Address'].str.replace(' ROAD', ' RD')
charlotte_df["Address"]= charlotte_df['Address'].str.replace(' LANE', ' LN')
charlotte_df["Address"]= charlotte_df['Address'].str.replace(' PLACE', ' PL')
# charlotte_df["Address"]= charlotte_df['Address'].str.replace(' ', '%')
```

```python
charlotte_df = charlotte_df.dropna(subset=['Address'])
charlotte_df.head(50)
```

```python
phone_num = []
for i in range(len(charlotte_df)):
    query_url = f'https://data.searchbug.com/api/search.aspx?CO_CODE=12617834&PASS=Ruger2012!?&TYPE=api_phone&ADDRESS={charlotte_df.Address[i]}&CITY=Charlotte&STATE=NC&FORMAT=JSON'
    response = requests.get(query_url)
    response2 = response.json()
    for x in range(len(response2) - 1):
        if response2['STATUS'] == 'NORESULTS':
            break
        else:
            try:
                response2['DATA']['RECORD'][x]['LNAME']
                if response2['STATUS'] == 'SUCCESS' and response2['DATA']['RECORD'][x]['LNAME'] == charlotte_df['Last Name'][i]:
                    phone_num.append(response2['DATA']['RECORD'][x]['PHONE'])
                    break
                else:
                    if response2['STATUS'] == 'SUCCESS' and response2['DATA']['RECORD'][(x + 1)]['LNAME'] == charlotte_df['Last Name'][i]:
                        phone_num.append(response2['DATA']['RECORD'][(x + 1)]['PHONE'])
                        break
                    else:
                        phone_num.append('NaN')
                        break
            except:
                if response2['DATA']['RECORD']['LNAME'] == charlotte_df['Last Name'][i]:
                    phone_num.append(response2['DATA']['RECORD']['PHONE'])
                else:
                    phone_num.append('NaN')
                    break
phone_num               
```
